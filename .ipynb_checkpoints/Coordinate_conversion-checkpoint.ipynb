```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
def to_coords(x, y, w, h, lat_centre, lon_centre, lat_range, lon_range):
    
    x = lon_centre - lon_range/2 + x*abs(lon_range )
    y = lat_centre - lat_range/2 + y*abs(lat_range )
    
    w = w*abs(lon_range)
    h = h*abs(lat_range)
    
    return x, y, w, h
```

```python
def plot_craters(x, y, w, h):
    
    if type(x) == int:
        x, y, w, h = [x], [y], [w], [h]
    
    for k in range(len(x)):
        ax.scatter(x[k], y[k])
        rect = patches.Rectangle((x[k] - w[k]/2, y[k] - h[k]/2), width=w[k], height=h[k], linewidth=1, edgecolor='r', facecolor='none')
        plt.gca().add_patch(rect)
```

```python
# define test data

lat_centre = np.arange(-45, 45, 10)
lon_centre = np.arange(-180, 0, 20)
lat_range = np.ones_like(lon_centre)*10
lon_range = np.ones_like(lon_centre)*10

x0 = np.random.randint(100, size =5)/100
y0 = np.random.randint(100, size =5)/100

w0 = np.random.randint(100, size =5)/100
h0 = np.random.randint(100, size =5)/100
```

```python
# plot test crater data for different latitude and longitude values: 

for i in range(len(lat_centre)):
    x, y, w, h = to_coords(x0, y0, w0, h0, lat_centre[i], lon_centre[i], lat_range[i], lon_range[i])
    fig, ax = plt.subplots(1,1, figsize = (3,3))
    plot_craters(x,y,w,h)
    
```
